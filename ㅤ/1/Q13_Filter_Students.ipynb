```
# Q13: Student Filtering
import pandas as pd

df = pd.DataFrame({
    "Name":["A","B","C","D","E","F"],
    "Age":[18,21,19,20,17,22],
    "Marks":[55,78,90,45,88,65]
})

avg = df["Marks"].mean()
print(df[df["Marks"]>avg])
print(df[(df["Age"]<20)&(df["Marks"]>60)]["Name"])
```
