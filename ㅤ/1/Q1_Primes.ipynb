# Q1 Primes

```python
# Q1: 1D Array, Primes, Mean & Variance
import numpy as np

arr = np.arange(1, 21)

def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

primes = np.array([x for x in arr if is_prime(x)])
print("Primes:", primes)
print("Mean:", primes.mean())
print("Variance:", primes.var())
```
