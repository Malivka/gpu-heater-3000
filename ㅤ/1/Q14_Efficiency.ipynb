```python
# Q14: Efficiency
import pandas as pd

df = pd.DataFrame({
    "Name":["A","B","C","D","E"],
    "Tasks_Completed":[20,25,18,30,22],
    "Hours_Worked":[5,6,4,8,5]
})

df["Efficiency"] = df["Tasks_Completed"]/df["Hours_Worked"]
best = df.loc[df["Efficiency"].idxmax()]
print(df)
print("Best:", best["Name"], best["Efficiency"])
```
