# MA3 PartB Q1 Bernoulli Binomial

```

# Q1: Bernoulli and Binomial Distribution

import numpy as np

# Bernoulli distribution
bernoulli = np.random.binomial(1, 0.6, 1000)
print("Bernoulli Mean:", np.mean(bernoulli))
print("Bernoulli Variance:", np.var(bernoulli))

# Binomial distribution
binomial = np.random.binomial(10, 0.5, 1000)
print("Binomial Mean:", np.mean(binomial))
print("Binomial Variance:", np.var(binomial))
```
