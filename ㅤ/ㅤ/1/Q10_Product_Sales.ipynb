```
# Q10: Product Sales
import pandas as pd

df = pd.DataFrame({
    "Name":["Pen","Book","Bag","Bottle","Shoes","Watch"],
    "Quantity":[10,5,3,8,2,4],
    "Price":[10,50,700,100,1200,1500]
})

df["Total"] = df["Quantity"]*df["Price"]
print(df)
print("Max Revenue:\n", df.loc[df["Total"].idxmax()])
```
