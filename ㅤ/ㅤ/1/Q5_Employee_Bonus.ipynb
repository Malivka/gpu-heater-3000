```
# Q5: Employee Bonus
import pandas as pd

df = pd.DataFrame({
    "ID":[1,2,3,4],
    "Name":["A","B","C","D"],
    "Salary":[30000,50000,40000,60000]
})

df["Bonus"] = df["Salary"]*0.10
avg = df["Salary"].mean()

print(df)
print("Above average:\n", df[df["Salary"]>avg])
```
