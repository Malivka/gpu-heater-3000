# Q3 Students Topper

```python
# Q3: Student DataFrame and Topper
import pandas as pd

df = pd.DataFrame({
    "Name":["A","B","C","D","E"],
    "Math":[78,85,90,66,88],
    "Science":[80,70,92,60,89],
    "English":[75,88,85,72,90]
})

df["Total"] = df[["Math","Science","English"]].sum(axis=1)
df["Average"] = df["Total"]/3

topper = df.loc[df["Average"].idxmax()]
print(df)
print("Topper:", topper["Name"], topper["Average"])
```
