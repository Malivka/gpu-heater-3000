```

# Q8: Binomial Distribution PMF, Mean, Variance, CDF

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 2000
p = 0.3

x = np.arange(0, 600)
pmf = binom.pmf(x, n, p)

plt.plot(x, pmf)
plt.title("Binomial PMF")
plt.show()

print("Mean:", n*p)
print("Variance:", n*p*(1-p))

cdf = binom.cdf(x, n, p)
plt.plot(x, cdf)
plt.title("Binomial CDF")
plt.show()
```
