```

# Q8: Encoding Categorical Data

import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.DataFrame({
    "Department":["HR","IT","Finance","HR","IT"],
    "Job_Role":["Manager","Analyst","Clerk","Analyst","Manager"],
    "Marital_Status":["Single","Married","Divorced","Single","Married"]
})

print("Original Data:\n", df)

# Label Encoding
le = LabelEncoder()
df["Marital_Status"] = le.fit_transform(df["Marital_Status"])

# One-hot Encoding
df_encoded = pd.get_dummies(df, columns=["Department","Job_Role"])

print("\nEncoded Data:\n", df_encoded)
```
