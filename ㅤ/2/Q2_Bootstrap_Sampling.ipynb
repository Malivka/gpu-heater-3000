```

# Q2: Bootstrap Sampling

import pandas as pd
from sklearn.utils import resample

df = pd.read_csv("btissue.csv")

# Feature columns only
X = df.iloc[:, :-1]

# Bootstrap sample
boot = resample(X, n_samples=100, replace=True, random_state=42)

print("First 10 rows of bootstrap sample:")
print(boot.head(10))

print("\nRepeated rows exist because sampling is done with replacement.")
```
